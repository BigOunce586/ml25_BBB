```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
ruta_csv = r"C:\Users\paola\Downloads\ml25_BBB\src\ml25\datasets\customer_purchases\customer_purchases_train_modify.csv"
dataFrame = pd.read_csv(ruta_csv)
dataFrame.head()


```

Distribución de Edad y Moda

```python

edad_moda = dataFrame['edad'].mode()[0]
plt.figure(figsize=(10, 6))
plt.hist(dataFrame['edad'], bins=30, edgecolor='black', alpha=0.7)
plt.axvline(x=edad_moda, color='r', linestyle='--', label=f'Moda: {edad_moda}')
plt.title('Distribución de Edad con Moda')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
category_map = {
    1.0: "jacket",
    0.6: "blouse",
    0.2: "jeans",
    -0.2: "skirt",
    -0.6: "shoes",
    -1.0: "dress"
}

color_map = {
    1: "azul",
    0.714: "negro",
    0.429: "verde",
    -0.429: "rosa",
    -0.714: "blanco",
    0.143: "naranja",
    -1: "amarillo",
    -0.1: "verde"
}

dataFrame["item_category"] = dataFrame["item_category"].map(category_map)
dataFrame["color"] = dataFrame["color"].map(color_map)
df_numeric = dataFrame[["edad", "customer_gender", "customer_signup_time", "item_price"]]
corr_matrix = df_numeric.corr()
plt.figure(figsize=(12, 8))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title('Matriz de Correlación entre Atributos Seleccionados')
plt.show()
```

```python
item_price_moda = dataFrame['item_price'].mode()[0]

plt.figure(figsize=(12, 6))
plt.hist(dataFrame['item_price'], bins=30, edgecolor='black', alpha=0.7)

plt.axvline(x=item_price_moda, color='r', linestyle='--', label=f'Moda: {item_price_moda:.2f}')

plt.title('Distribución del Precio del Producto con Moda')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.legend()

plt.show()
```

```python
gender_moda = dataFrame['customer_gender'].mode()[0]
plt.figure(figsize=(10, 6))
plt.hist(dataFrame['customer_gender'], bins=30, edgecolor='black', alpha=0.7)

plt.axvline(x=gender_moda, color='r', linestyle='--', label=f'Moda: {gender_moda}')
plt.title('Distribución de genero con Moda')
plt.xlabel('genero')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
print("1 es female , 2 es male ,3 es null")
```
